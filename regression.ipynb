```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("D:/Program Files (x86)/New folder/Housing.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
num_values=df.select_dtypes(include=['int64','float64'])
num_values.head()
```

```python
cat_values=df.select_dtypes(include=['object','category','bool'])
cat_values.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
cat_values_le=cat_values.apply(le.fit_transform)
cat_values_le.head()
```

```python
data=pd.concat([num_values,cat_values_le],axis=1)
data.head()
```

```python
y=data["price"]
x=data.drop(labels="price",axis=1)
```

# Linear Regression

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.80,random_state=100)
```

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
temp_df=pd.DataFrame({'Actual':y_test,'predicted':y_pred})
temp_df.head()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```
